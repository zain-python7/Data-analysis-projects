```python
# Import Matplotlib, pandas, and plotly
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

**Task 1.5.1** 

```python
df1 = pd.read_csv("data/brasil-real-estate-1.csv")
df1.head()
```

**Task 1.5.2** 

```python
df1.dropna(inplace=True)
df1.head()
```

**Task 1.5.3** 

```python
df1[["lat", "lon"]] = df1["lat-lon"].str.split(",",expand=True)
df1["lat"]=df1["lat"].astype(float)
df1["lon"]=df1["lon"].astype(float)
df1.head()
```

```python

df1.head()
```

**Task 1.5.4** 

```python
df1["state"] = df1["place_with_parent_names"].str.split("|",expand=True)[2]
df1.head()
```

**Task 1.5.5** 

```python
df1["price_usd"] =df1["price_usd"].str.replace("$","",regex=False).str.replace(",","",regex=False)
df1["price_usd"]=df1["price_usd"].astype(float)
df1.head()
```

```python

df1.head()
```

**Task 1.5.6** 

```python
df1=df1.drop(columns="lat-lon")
df1=df1.drop(columns="place_with_parent_names")
df1.head()
```

**Task 1.5.7** 

```python
df2 =pd.read_csv("data/brasil-real-estate-2.csv")
df2.head()
```

**Task 1.5.8** 

```python
df2["price_usd"] = df2["price_brl"]/3.19
df2.head()
```

**Task 1.5.9** 

```python
df2.dropna(inplace=True)
df2=df2.drop(columns="price_brl")
df2.head()
```

**Task 1.5.10** 

```python
df =pd.concat([df1,df2])
print("df shape:", df.shape)
df.head()
```

**Task 1.5.12** 

```python
df=df.head(20000)
# Don't change the code below 👇
fig, ax = plt.subplots()

# Build histogram using plt
ax.hist(df["price_usd"])

# Label axes
ax.set_xlabel("Price [USD]")
ax.set_ylabel("Frequency")

# Add title
ax.set_title("Distribution of Home Prices");
```

**Task 1.5.13** 

```python
# Don't change the code below 👇
fig, ax = plt.subplots()

#Build box plot
ax.boxplot(df["area_m2"],vert=False)

# Label x-axis
ax.set_xlabel("Area [sq meters]")
# Add title
ax.set_title("Distribution of Home Sizes");
```

**Task 1.5.14** 

```python
mean_price_by_region =df.groupby("region")["price_usd"].mean().sort_values(ascending=True)
mean_price_by_region
```

**Task 1.5.15** 

```python
# Don't change the code below 👇
fig, ax = plt.subplots()


# Build bar chart, label axes, add title
mean_price_by_region.plot(
    kind="bar",
    xlabel="Region",
    ylabel="Mean Price [USD]",
    title="Mean Home Price by Region",
    ax=ax);
```

**Task 1.5.16**

```python
df_south = df[df["region"]=="South"]
df_south.head()
```

**Task 1.5.17** 

```python
homes_by_state =df_south["state"].value_counts()

homes_by_state
```

**Task 1.5.18** 

```python



most_common_state = df_south["state"].value_counts().idxmax()


df_south_rgs = df_south[df_south["state"] == most_common_state]


fig, ax = plt.subplots()

ax.scatter(df_south_rgs["area_m2"], df_south_rgs["price_usd"])

# Step 4: Label axes and title with actual state name
ax.set_xlabel("Area [sq meters]")
ax.set_ylabel("Price [USD]")
ax.set_title(f"{most_common_state}: Price vs. Area")

plt.show()
```

**Task 1.5.19** 

```python

south_states_corr = {}
for south in df_south["state"].unique():
    df_state=df_south[df_south["state"]==south]
    corr=df_state["area_m2"].corr(df_state["price_usd"])
    south_states_corr[south]=corr
    

south_states_corr
```
